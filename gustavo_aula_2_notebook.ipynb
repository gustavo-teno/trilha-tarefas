## Questão 1 📝

Escreva um programa que leia N números e exiba uma mensagem informando se cada número é positivo, negativo ou zero.

Use um loop infinito com condição de parada: quando o usuário digitar `"pare"`, o programa encerra.

````python

while True:
  entrada = input()

  if entrada == "pare":
    break

  else:
    entrada = float(entrada)
    if entrada > 0:
      print("positivo")
    elif entrada < 0:
      print("negativo")
    else:
      print("zero")

      #obs: converti para float ao invés de int
````

## Questão 2 📝

Faça um programa que peça ao usuário para digitar senhas até que ele acerte a senha correta.

A senha predefinida é: `"trilhaaula2"`

Após acertar, mostre quantas tentativas incorretas foram necessárias.

````python
senha_correta = "trilhaaula2"
tentativas = 0
count = 0

while count == 0:
  entrada = input()
  tentativas += 1

  if entrada != senha_correta:
    continue
  else:
    #apenas tentativas incorretas
    print(f"\n{tentativas - 1}")
    count = 1
    break
````

## Questão 3 📝

Crie um programa que peça um número ao usuário e determine se ele é primo.

*Lembrete: número primo é divisível apenas por 1 e por ele mesmo. O número 1 **não** é primo.*

````python
sim = "Primo"
nao = "Não é primo"

while True:
  entrada = input()

  if entrada.lower() == "pare":
    break

  entrada = int(entrada)

  if entrada % 2 == 0 and entrada != 2 or entrada == 1:
    print(nao)

  else:
    eh_primo = True
    limite = round(entrada**0.5)

    for i in range(2,limite+1):
      if entrada % i == 0:
        eh_primo = False
        break

    if eh_primo:
      print(sim)
    else:
      print(nao)
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Use `%` para testar divisores entre 2 e `numero - 1`. Encontrou algum? Não é primo. Chegou ao fim sem encontrar? É primo.*

*Curiosidade: o bloco `else` de um `for` só executa quando o loop termina **sem** `break`.*

## Questão 4 📝

Escreva um programa que exiba a tabuada completa de 1 a 10.

**Exemplo de saída esperada (trecho):**
```
Tabuada do 1:
1 x 1 = 1
1 x 2 = 2
...
```

````python
numero = int(input("Digite um número:\n"))
print(f"Tabuada do {numero}:")

count = 1
while count <= 10:
  produto = numero * count
  print(f"{numero} x {count} = {produto}")
  count += 1
````

## Questão 5 📝

Crie uma função chamada `calcular_media(n1, n2, n3)` que recebe três notas, calcula a média e **retorna** o resultado.

Depois, chame a função com três valores e imprima a média usando f-string com 2 casas decimais.

````python
def calcular_media(n1: float, n2: float, n3: float) -> float:
  resultado = (n1 + n2 + n3)/3
  return resultado

media = calcular_media(10.00, 4.00, 7.00)
print(f"Média = {media:.2f}")
````

## Questão 6 📝

Crie uma função chamada `par_ou_impar(numero)` que recebe um número inteiro e **imprime** se ele é par ou ímpar.

*Atenção: esta função não precisa retornar nada.*

Depois, chame a função com 3 números diferentes.

````python
def par_ou_impar(numero: int) -> None:
  if numero % 2 == 0:
    print("Par")
  else:
    print("Ímpar")

par_ou_impar(5)
par_ou_impar(42)
par_ou_impar(101)
````

## Questão 7 📝

Crie uma função chamada `calcular_fatorial(n)` que recebe um número inteiro positivo e **retorna** seu fatorial usando um loop `for`.

*Lembrete: fatorial de n = 1 × 2 × 3 × ... × n. Por convenção, fatorial de 0 é 1.*

Depois, chame a função para calcular o fatorial de 5 e imprima o resultado.

````python
def calcular_fatorial(n: int) -> int:
  #fator um -> fatorial | valor inicial = 1
  fat = 1

  #fator dois -> i, que a cada iteração vai aumentar o seu valor em +1
  for i in range(1, n+1):
    fat *= i

  return fat

print(calcular_fatorial(5))
````

### 💡 Dica:

*Utilize somente se necessário — tente resolver primeiro sozinho!*

*Inicie `resultado = 1` (elemento neutro da multiplicação) e vá multiplicando a cada iteração. Qual `range()` percorre de 1 até n **inclusive**?*
